```python
import numpy as np
from tensorflow.keras.applications import VGG16, imagenet_utils
from tensorflow.keras.preprocessing.image import load_img, img_to_array
import matplotlib.pyplot as plt

# Load pre-trained model
model = VGG16(weights="imagenet")

# Load and preprocess image
image_path = "/content/AAI(8).jpg" # Replace with your image path
image = load_img(image_path, target_size=(224, 224))
image_array = img_to_array(image)
image_batch = np.expand_dims(image_array, axis=0)
image_preprocessed = imagenet_utils.preprocess_input(image_batch.copy())

# Predict
preds = model.predict(image_preprocessed)
decoded_preds = imagenet_utils.decode_predictions(preds, top=3)[0]

# Display image
plt.imshow(image_array.astype("uint8"))
plt.axis("off")
plt.title("Input Image")
plt.show()

# Textual explanation
top_label, top_desc, top_conf = decoded_preds[0]
print("\nTextual Explanation:\n")
print(f"The model predicts this image as a '{top_desc}' with a confidence of {top_conf * 100:.12f}%.")
print(f"This prediction is based on patterns in the image that strongly match features commonly found in '{top_desc}' images from the ImageNet dataset.")
print("Other possible predictions include:")
for i, (_, label, prob) in enumerate(decoded_preds[1:], start=2):
    print(f"  {i}. {label} ({prob * 100:.12f}%)")
```
